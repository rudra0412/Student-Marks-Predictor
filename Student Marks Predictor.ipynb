```python
#Importing essential libraries

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Reading the data from remote link

url="http://bit.ly/w-data"
df=pd.read_csv(url)
df.head()
```

```python
#getting the shape of the data sample
df.shape
```

```python
#plotting a scatter plot of distribution of scores

plt.scatter(x=df.Hours, y=df.Scores)
plt.title('Progress Report')
plt.xlabel('Study Hours')
plt.ylabel('Student Marks')
plt.show()
```

```python
#Identifying the missing values

df.isnull().sum()
```

```python
#Separating dependent and independent variables

X=df.drop("Scores", axis="columns")
y=df.drop("Hours", axis="columns")
```

```python
#Splitting the data into training and testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Training the linear regression model on out data

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Getting the input from user for testing

sh=float(input("Enter Study hours"))
print('Score: ',lr.predict([[sh]])[0][0].round(2))
```

```python
#Graphical representation of the solution

y_pred=lr.predict(X_test)

plt.scatter(x=df.Hours, y=df.Scores, color='blue')
plt.plot(X_test,y_pred,color='red') #linear regression line
plt.title('Progress Report')
plt.xlabel('Study Hours')
plt.ylabel('Student Marks')
plt.show()
```

```python
#present your solution
import joblib
joblib.dump(lr, 'marks.pkl')
```
